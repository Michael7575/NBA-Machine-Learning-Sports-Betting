```python
import pandas as pd
```

```python
raw_df=pd.read_csv('Datasets/Full-Data-Set-UnderOver-2021-22.csv')
```

## Clean data 

```python
raw_df.drop(['Win_Margin', 'Unnamed: 0'], axis=1, errors='ignore',inplace=True)
```

```python
raw_df.columns=raw_df.columns.str.lower()
```

```python
raw_df=raw_df[~raw_df['ml_home'].isin(["NL"])]
```

```python
raw_df['ml_home']=raw_df['ml_home'].astype(float)
raw_df['ml_away']=raw_df['ml_away'].astype(float)
```

```python
raw_df.columns
```

## Split data

```python
raw_df['date'] = pd.to_datetime(raw_df['date'])
```

```python
validation_cutoff_dt = '2020-10-01'
```

```python
train_df = raw_df[raw_df['date']<validation_cutoff_dt]
vald_df = raw_df[raw_df['date']>=validation_cutoff_dt]
```

## Train Win/Loss model

```python
from pycaret.classification import *
```

```python
list(train_df.columns)
```

```python
exp_name = setup(data = train_df, 
                 target = 'home-team-win',
                 test_data = vald_df,
                 ignore_features = [ 'home','away','ou-cover','ou'],
#                  ordinal_features = ordinal_dicts,
                 feature_selection = True,
                 feature_selection_threshold = 0.9,
                 remove_multicollinearity = True,
                 multicollinearity_threshold = 0.9,
                 normalize=True)
```

```python
# eda(display_format = 'bokeh')
```

```python
#Feature selection + multicollinearity
# top3 = compare_models(include=['lr','lda','gbc'], n_select = 3)
```

```python
lda=create_model('lda')
```

```python
tuned_lda=tune_model(lda)
```

```python
evaluate_model(tuned_lda)
```

```python
final_lda = finalize_model(tuned_lda)
```

```python
save_model(final_lda, 'win_loss_acc_72')
```
